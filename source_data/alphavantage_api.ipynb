```python
import numpy as np
import pandas as pd
import requests
from datetime import datetime, timedelta
```

```python
key = 'JCIBBY8DVEGEOYO'
```

```python
# Getting Inflation rate from Alpha Vantage

url = f'https://www.alphavantage.co/query?function=INFLATION&interval=monthly&apikey={key}'
response = requests.get(url).json()

#converting to dataframe
df = pd.json_normalize(response, record_path =['data'])

#converting to columns to datetime
df['date'] = pd.to_datetime(df['date']).dt.normalize()
df['fy'] = df['date'].dt.year
df = df.loc[df['fy'] >= 1980]
df = df.drop(labels=['date'], axis=1)
df = df.rename(columns={'value' : 'inflation'})
df['inflation'] = df['inflation'].astype('float').round(2)
inflation = df.copy()
inflation.head()
```

```python
# Getting Inflation rate from Alpha Vantage

url = f'https://www.alphavantage.co/query?function=UNEMPLOYMENT&interval=monthly&apikey={key}'
response = requests.get(url).json()

#converting to dataframe
df = pd.json_normalize(response, record_path =['data'])

#converting to columns to datetime
df['date'] = pd.to_datetime(df['date']).dt.normalize()
df['fy'] = df['date'].dt.year
df = df.loc[df['fy'] >= 1980]
df = df.drop(labels=['date'], axis=1)
df = df.rename(columns={'value' : 'unemployment'})
df['unemployment'] = df['unemployment'].astype('float').round(2)
unemployment = df.copy()
unemployment.head()
```

```python
# Getting Interest rate from Alpha Vantage

url = f'https://www.alphavantage.co/query?function=FEDERAL_FUNDS_RATE&interval=monthly&apikey={key}'
response = requests.get(url).json()

#converting to dataframe
df = pd.json_normalize(response, record_path =['data'])

#converting to columns to datetime
df['date'] = pd.to_datetime(df['date']).dt.normalize()
df['fy'] = df['date'].dt.year
df = df.loc[df['fy'] >= 1980]
df = df.drop(labels=['date'], axis=1)
df = df.rename(columns={'value' : 'interest'})
df['interest'] = df['interest'].astype('float').round(2)
interest = df.copy()
interest.head()
```

```python
final = pd.merge(interest, unemployment, on='fy')
final = pd.merge(final, inflation, on='fy')
final
```

```python
final.to_csv('econ.csv')
```
